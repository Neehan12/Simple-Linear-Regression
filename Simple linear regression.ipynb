# Simple linear regression for beginners

Importing the necessary libraries :


```python
import pandas as pd
import statsmodels.formula.api as smf
```

Importing the data through pandas library that we just imported in the above line of code

```python
pd.read_csv('Salary_data.csv') 
```

```python
#Storing the dataset(salary_data.csv) in a variable(df) for further use
df = pd.read_csv('Salary_data.csv')
df
```

```python
#Since in SIMPLE linear regression, we need one feature and one target , we'll be using one column as feature(x) and other column as target(y)
x=df['YearsExperience'] 
y=df['Salary'] 
```

```python
#Now we fit these variables x and y into the simple linear regressor model with smf
model = smf.ols("y~x",data=df).fit()
```

```python
#We check the accuracy of the model now ( The accuracy tells us how good the model is at predicting the Salary of a person based on his/her years of experience)
model.rsquared , model.rsquared_adj
print('These scores show that accuracy of the model is 95% which is extremely good')
print (model.rsquared , model.rsquared_adj)
#The print function is used to print something under the cell as you can see.
```

# Now we know how to fit data into a simple linear regressor model, We also got good accuracy, but what to do if your data does not show good results? how to improve accuracy for your simple linear regressor model? Lets See !

```python
model1 = smf.ols("Y~X" ,data =df2).fit()
```

```python
print(model1.rsquared, model1.rsquared_adj)
print('Now as you see, the Accucary is 68%, which is not a very good number, now lets try to improve it')
```

R2 Score is the score that tells us how accurate the model is, to improve the accuracy of a model, there are various techniques. We will be trying a few BASIC SCALING METHODS. To learn more about what Scaling is, check out my notebooks on Scaling :D

## For Scaling the data, We'll need to import a library that helps us apply scaling functions on the data, so here we go!

```python
import numpy as np
df3 = np.log(df2)
df3
print( 'So basically We are scaling the data by taking a log of it, This bring the uneven data consisting of large and small numbers into a consistent range so the model could understand it better. lets see our results now' )
```

```python
x2=df3.iloc[:,-1] # Feature
y2=df3.iloc[:,0] # Target
model=smf.ols("y2~x2",data=df3).fit()
print(model.rsquared, model.rsquared_adj)
print('As you can see, Our model accuracy has improved from 68% to 77%. This was one of the scaling techniques used to improve accuracy, You can learn about more of them in one of my notebooks on Scaling')
```
